# Simple One Hidden Layer Neural Network


```python
import torch
import torch.nn as nn
from torch import sigmoid
import matplotlib.pylab as plt

%matplotlib inline
torch.manual_seed(0)
```

```python
def PlotStuff(X, Y, model, epoch, leg=True):
    plt.plot(X.numpy(), model(X).detach().numpy(), label=("epoch " + str(epoch)))
    plt.plot(X.numpy(), Y.numpy(), "r")
    plt.xlabel("x")
    if leg == True:
        plt.legend()
    else:
        pass
```

```python
class Net(nn.Module):
    def __init__(self, D_in, H, D_out):
        super(Net, self).__init__()
        # hidden layer
        self.linear1 = nn.Linear(D_in, H)
        self.linear2 = nn.Linear(H, D_out)
        # Define the first linear layer as an attribute, this is not good practice
        self.a1 = None
        self.l1 = None
        self.l2 = None

    # Prediction
    def forward(self, x):
        self.l1 = self.linear1(x)
        self.a1 = sigmoid(self.l1)
        self.l2 = self.linear2(self.a1)
        yhat = sigmoid(self.l2)
        return yhat
```

```python
def train(Y, X, model, optimizer, criterion, epochs=1000):
    cost = []
    total = 0
    for epoch in range(epochs):
        total = 0
        for y, x in zip(Y, X):
            yhat = model(x)
            loss = criterion(yhat, y)
            loss.backward()
            optimizer.step()
            optimizer.zero_grad()
            # cumulative loss
            total += loss.item()
        cost.append(total)
        if epoch % 300 == 0:
            PlotStuff(X, Y, model, epoch, leg=True)
            plt.show()
            model(X)
            plt.scatter(
                model.a1.detach().numpy()[:, 0],
                model.a1.detach().numpy()[:, 1],
                c=Y.numpy().reshape(-1),
            )
            plt.title("activations")
            plt.show()
    return cost
```

```python
# Make Data
X = torch.arange(-20, 20, 1).view(-1, 1).type(torch.FloatTensor)
Y = torch.zeros(X.shape[0])
Y[(X[:, 0] > -4) & (X[:, 0] < 4)] = 1.0
```

```python
def criterion_cross(outputs, labels):
    out = -1 * torch.mean(
        labels * torch.log(outputs) + (1 - labels) * torch.log(1 - outputs)
    )
    return out
```

```python
D_in = 1  # size of input
H = 2  # size of hidden layer
D_out = 1  # number of outputs
learning_rate = 0.1  # learning rate

model = Net(D_in, H, D_out)
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
cost_cross = train(Y, X, model, optimizer, criterion_cross, epochs=1000)

plt.plot(cost_cross)
plt.xlabel("epoch")
plt.title("cross entropy loss")
```

```python
# Using MSE Loss to train the same model

learning_rate = 0.1
criterion_mse = nn.MSELoss()
model = Net(D_in, H, D_out)
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
cost_mse = train(Y, X, model, optimizer, criterion_mse, epochs=1000)

plt.plot(cost_mse)
plt.xlabel("epoch")
plt.title("MSE loss ")
```
